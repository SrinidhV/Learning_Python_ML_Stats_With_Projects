To remove duplicates


```python
df = df.drop_duplicates()
```

```python
df
```

To drop a column 

```python
df = df.drop(columns = ['Not_Useful_Column'])
df
```

Strip Unnecessary Characters

```python
df['Last_Name'] = df['Last_Name'].str.strip() # By default it removes only white spaces
df['Last_Name'] = df['Last_Name'].str.lstrip(".") # Removes the ... from left side of the string
df['Last_Name'] = df['Last_Name'].str.rstrip("_")
df['Last_Name'] = df['Last_Name'].str.strip("/")

# Strip works like a RE. We can directly remove all of the stuff by mentioning in a single paranthesis

df['Last_Name'] = df['Last_Name'].str.strip("/._ ") # This alone does the work of all the above 4 statements
df
```

Using regex to replace stuff

```python
#Before we use .str and regex to replace stuff, we need everything to be a string in that column
df['Phone_Number'] = df['Phone_Number'].astype('string')

df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]' , '', regex = True).fillna('')
df['Phone_Number'].apply(type).value_counts()

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
df['Phone_Number'] = df['Phone_Number'].replace('--', '')

df
```

Splitting Columns

```python
df[["Street","State","Zipcode"]]= df['Address'].str.split(pat=',', n=2, expand=True)
df
```

```python
df[['Paying Customer','Do_Not_Contact']] = df[['Paying Customer','Do_Not_Contact']].replace(['Y','N','N/a'], ['Yes','No','']).fillna('')
df
```

```python

df= df.fillna('') # Nan And None value's are considered as na
df
```

To drop rows based on a filter. We can also we for loop

```python
a = df[df['Do_Not_Contact'] == 'Yes'].index
b= df[df['Phone_Number'] == ''].index
df.drop(a,inplace= True)
df.drop(b,inplace= True)
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('', 'No')
df.drop(['Address'], axis = 1)
df.reset_index(drop  =True)
```
